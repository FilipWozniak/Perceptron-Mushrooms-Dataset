```python
from sklearn import datasets
import numpy as np
import pandas as pd
import math
import random
```

```python
class Perceptron(object):
    def __init__(self, columns_number, iterations = 30, threshold = 1.0, learning_rate = 0.1):
        self.iterations = iterations
        self.threshold = threshold
        self.learning_rate = learning_rate
        self.weights = np.zeros(columns_number)
        
    #   WX = (X1*W1) + (X2*W2) + .. + (XN*WN)
    def predict(self, row):
        activation = 0.0
        activation = np.dot(self.weights, row)
        return 1.0 if activation >= self.threshold else 0.0
    
    def train_weights(self, training_set):
        for iteration in range(self.iterations):
            for index, row in training_set.iterrows():
                expected = row[0]
                row = row.iloc[1:]
                a = np.array(row.values.tolist())
                prediction = self.predict(a)
                error = expected - prediction                
                self.weights += self.learning_rate * error * a
        return self.weights
```

```python
# READ THE .CSV FILE
data_mushrooms = pd.read_csv(r"mushrooms.csv")
data_mushrooms.head(5)
```

```python
print("YOU CANNOT SPECIFY MORE THAN " + str(len(data_mushrooms.columns)-1))
number_of_columns = int(input("SPECIFY NUMBER OF COLUMNS: "))
```

```python
if (number_of_columns > len(data_mushrooms.columns)-1) or (number_of_columns < 2):
    raise ValueError("Error")
```

```python
# SLICE AN ARRAY (USE PRE-DEFINED BY USER NUMBER OF COLUMNS)
# IT MAY TAKE A WHILE
i = 0
for column in data_mushrooms:
    if i > number_of_columns:
        data_mushrooms = data_mushrooms.drop(column, axis=1)
    i = i + 1

# MAP ALL THE VALUES TO NUMBERS
def map_to_numbers():
    for column in data_mushrooms:
        helper = []
        for index, row in data_mushrooms.iterrows():
            helper.append(row[column])
        map_to_numbers = dict([(y, x + 1) for x, y in enumerate(sorted(set(helper)))])
        data_mushrooms[column] = data_mushrooms[column].map(map_to_numbers)
    return data_mushrooms

data_mushrooms_training_set = map_to_numbers()
data_mushrooms_training_set.head(5)
```

```python
# CREATE TESTING SET (A SET BETWEEN 100 ROWS AND 1/4 ROWS OF THE TRAINING SET)
bound_down = 100
bound_up = data_mushrooms_training_set['class'].count() / 4
data_mushrooms_testing_set = data_mushrooms_training_set.sample(n=random.randint(bound_down, bound_up))
```

```python
for i in data_mushrooms_testing_set.index:
    data_mushrooms_testing_set.at[i, 'class'] -= 1
for i in data_mushrooms_training_set.index:
    data_mushrooms_training_set.at[i, 'class'] -= 1

data_mushrooms_training_set.head(5)
```

```python
perceptron = Perceptron(columns_number = number_of_columns)
```

```python
# IT MAY TAKE A WHILE
perceptron.train_weights(data_mushrooms_training_set)
```

```python
predictions = list()
counter = 0
success = 0
for index, row in data_mushrooms_testing_set.iterrows():
    c = np.array(row.values.tolist())
    prediction = perceptron.predict(c[1:])
    predictions.append(int(prediction))
    if(prediction == row['class']):
        success += 1
    counter += 1
print('Successes:', success, '| Fails:', counter - success, f'| {math.ceil(success / counter * 100)}%')
```
